# Limits of Diversification

```python
%load_ext autoreload
%autoreload 2
import edhec_risk_kit as erk
import pandas as pd
import numpy as np
```

```python
ind_return = erk.get_ind_returns()
ind_nfirms = erk.get_ind_nfirms()
ind_size = erk.get_ind_size()
```

```python
ind_return.shape
```

```python
ind_nfirms.shape
```

```python
ind_size.shape
```

```python
ind_nfirms.head()
```

```python
ind_mktcap = ind_nfirms * ind_size
ind_mktcap.shape
```

```python
total_mktcap = ind_mktcap.sum(axis=1)
```

```python
total_mktcap.plot(figsize=(12,6))
```

```python
ind_capweihght = ind_mktcap.divide(total_mktcap, axis=0)
```

```python
ind_capweihght.head()
```

```python
ind_capweihght["1926":"1926"].sum(axis=1)
```

```python
ind_capweihght[["Fin", "Steel"]].plot(figsize=(12,6))
```

```python
total_market_return = (ind_capweihght * ind_return).sum(axis=1)
```

```python
total_market_return.plot()
```

```python
total_market_index = erk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize=(10, 5), title="Total Market CapWeighted Index")
```

```python
total_market_index["1980":].plot(figsize=(12,6))
total_market_index["1980":].rolling(window=36).mean().plot()
```

```python
tmi_tr36mrets = total_market_return.rolling(window=36).aggregate(erk.annualize_rets, periods_per_year=12)
tmi_tr36mrets.plot(figsize=(12,6), label="Tr36 mo Return", legend=True)
# total_market_return.plot(label="Returns", legend=True)
```

### Rolling Correlation - along with MultiIndexes and `.groupby`

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.tail()
```

```python
ind_return.head()
```

```python
ts_corr.index.names = ['date', 'industry']
ts_corr.tail()
```

```python
ind_tr36corr = ts_corr.groupby(level='date').apply(lambda cormat: cormat.values.mean())
```

```python
ind_tr36corr.plot()
```

```python
tmi_tr36mrets.plot(label="Tr36 Mo Rets", figsize=(12,6), legend=True)
ind_tr36corr.plot(label="Tr36 Mo Corr", legend=True, secondary_y=True)
```

```python
tmi_tr36mrets.corr(ind_tr36corr)
```
